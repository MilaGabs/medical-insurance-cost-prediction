## **1. Analise e ´pré-processamento**

A base de dados utilizada neste estudo foi obtida do site Kaggle.

Na análise inicial, examinamos a estrutura da base de dados, identificando as colunas existentes e calculando estatísticas descritivas como tendência central, dispersão, forma da distribuição e a presença de valores nulos. Verificamos que não havia valores nulos, portanto, não foi necessário realizar tratamento para dados ausentes.

Em seguida, realizamos uma análise gráfica para entender a distribuição dos dados. Geramos histogramas para as variáveis idade, IMC, número de filhos e encargos, e observamos uma tendência decrescente geral nas variáveis analisadas. Em gráficos de pizza, a distribuição estava equilibrada, exceto para a coluna de fumantes, que apresentava uma menor proporção de fumantes.

Criamos uma cópia da base de dados, denominada correlation_dataset, para gerar um heatmap e explorar a correlação entre as colunas. Identificamos que as variáveis "fumante" e "encargos" e "idade" e "encargos" mostraram correlações notáveis. Além disso, aplicamos o LabelEncoder para as colunas gênero, fumante e região, a fim de preparar os dados para a modelagem subsequente.

Para investigar a relação entre encargos e status de fumante, geramos um gráfico de barras mostrando a média de encargos por faixa etária para fumantes e não fumantes, observando uma diferença significativa entre os grupos em todas as idades.

Analisamos o equilíbrio entre os gêneros feminino e masculino com um pairgrid do Seaborn, destacando o gênero. Identificamos alguns outliers que poderiam impactar o desempenho do modelo. Utilizamos o zscore com um limiar de 3 para remover valores extremos e pacificar os dados. Após a remoção dos outliers, confirmamos a eficácia do tratamento gerando um gráfico adicional para a coluna de encargos.

Com os outliers removidos, calculamos a média de encargos por gênero e região, que foram adicionados ao dataset após a aplicação de one-hot encoding. Além disso, arredondamos os dados com casas decimais para duas casas após a vírgula.

## **2. Modelagem**

### **2.1. Treino e Teste**

- **Variável Alvo:** encargos (`y`);
- **Outras Variáveis:** idade, imc, filhos,	genero_feminino, genero_masculino, regiao_nordeste, regiao_noroeste, regiao_sudeste, regiao_sudoeste, fumante_não, fumante_sim (`x`);

Dividimos os dados em conjuntos de treinamento e teste da seguinte forma:

- **Treinamento:** 75% dos dados
- **Teste:** 25% dos dados

### **2.2. Modelo Utilizado**

Para selecionar o melhor modelo, implementamos uma função de validação cruzada, avaliando os modelos LinearRegression, KNeighborsRegressor, SVR e RandomForestRegressor, todos com hiperparâmetros básicos para garantir um equilíbrio na avaliação. O modelo com o melhor desempenho foi o RandomForestRegressor. Em seguida, otimizamos os hiperparâmetros usando RandomizedSearchCV para encontrar os melhores parâmetros e aplicá-los ao nosso modelo.
